Tâche 1: Collecte des données

```python
import os
import json
import requests
from dotenv import load_dotenv
import shutil
from PIL import Image
from PIL.ExifTags import TAGS

# 1. Chargement des variables d'environnement
load_dotenv('.env.local')
ACCESS_KEY = os.getenv('ACCESS_KEY')

# Nettoyage du dossier data au début de la tâche 1
DATA_DIR = "data"
os.makedirs(DATA_DIR, exist_ok=True)
for fname in os.listdir(DATA_DIR):
    fpath = os.path.join(DATA_DIR, fname)
    if os.path.isfile(fpath):
        os.remove(fpath)

METADATA_PATH = "data/images_metadata.json"


def download_metadata(photo, path, url, topic=None):
    """Enregistre les métadonnées de l'image dans data/images_metadata.json."""
    os.makedirs("data", exist_ok=True)

    # Charger les métadonnées existantes ou initialiser une liste vide
    metadata_list = []
    if os.path.exists(METADATA_PATH):
        try:
            with open(METADATA_PATH, "r", encoding="utf-8") as f:
                content = f.read().strip()
                if content:
                    metadata_list = json.loads(content)
        except (json.JSONDecodeError, ValueError):
            metadata_list = []

    filename = os.path.basename(path)
    file_size_kb = round(os.path.getsize(path) / 1024, 2)

    # Dimensions et format avec PIL
    try:
        with Image.open(path) as img:
            width, height = img.size
            file_format = img.format or os.path.splitext(filename)[1].lstrip(".").upper()
            exif_data = {}
            if hasattr(img, "getexif") and img.getexif():
                for tag_id, value in img.getexif().items():
                    tag_name = TAGS.get(tag_id, tag_id)
                    try:
                        exif_data[str(tag_name)] = str(value)
                    except (TypeError, ValueError):
                        pass
    except Exception:
        width, height = None, None
        file_format = os.path.splitext(filename)[1].lstrip(".").upper()
        exif_data = {}

    entry = {
        "file_name": filename,
        "topic": topic,
        "dimensions": {"width": width or photo.get("width"), "height": height or photo.get("height")},
        "file_format": file_format,
        "size_ko": file_size_kb,
        "url_source": url,
        "informations_licence": "Non spécifié",
        "exif": exif_data if exif_data else None,
        "created_at": photo.get("created_at"),
        "description": photo.get("description"),
        "username": photo.get("user", {}).get("username") if photo.get("user") else None,
    }
    metadata_list.append(entry)

    with open(METADATA_PATH, "w", encoding="utf-8") as f:
        json.dump(metadata_list, f, ensure_ascii=False, indent=2)


def download_image(photo, i, topic=None):
    # 1. Créer le dossier 'images' s'il n'existe pas déjà
    url = photo['urls']['regular']
    os.makedirs("images", exist_ok=True)

    headers = {"User-Agent": "Mozilla/5.0"}
    request = requests.get(url, allow_redirects=True, headers=headers, stream=True)
    
    if request.status_code == 200:
        # 2. Nettoyer le nom du fichier (avant le '?') et ajouter .jpg
        filename = str(i + 1) + ".jpg"
        
        # 3. Créer le chemin complet vers le dossier images
        filepath = os.path.join("images", filename)
        
        with open(filepath, "wb") as image:
            request.raw.decode_content = True
            shutil.copyfileobj(request.raw, image)

        download_metadata(photo, filepath, url, topic)
            
    return request.status_code
        
def fetch_100_random_photos(query):
    url = "https://api.unsplash.com/search/photos"
    headers = {
        "Authorization": f"Client-ID {ACCESS_KEY}"
    }
    
    total_needed = 100
    per_page = 10
    n_pages = (total_needed + per_page - 1) // per_page  # ici 10 pages de 10
    
    print(f"Début de la récupération de {total_needed} photos...")

    for page in range(1, n_pages + 1):
        params = {
            "query": query,
            "page": page,
            "per_page": per_page,
            "orientation": "landscape"
        }

        response = requests.get(url, headers=headers, params=params)
        response.raise_for_status()  # Lève une erreur si la requête échoue
        
        photos_data = response.json()["results"]
        
        for idx, photo in enumerate(photos_data):
            global_idx = (page - 1) * per_page + idx
            if global_idx >= total_needed:
                break
            download_image(photo, global_idx, query)
            
        done = min(page * per_page, total_needed)
        print(f"Progression : {done}/{total_needed}")

    print("Opération terminée.")

if __name__ == "__main__":
    query = input("Quel mot clef voulez vous utiliser pour chercher vos photos")
    fetch_100_random_photos(query)
```

Tâche 2: Étiquetage et annotation

```python
import os
import json
import re
from collections import Counter
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

LABELS_PATH = "data/images_labels.json"

# Palette simplifiée pour nommer les couleurs dominantes
BASE_COLORS = {
    "noir": (0, 0, 0),
    "blanc": (255, 255, 255),
    "gris": (128, 128, 128),
    "rouge": (220, 20, 60),
    "vert": (34, 139, 34),
    "bleu": (30, 144, 255),
    "jaune": (255, 215, 0),
    "orange": (255, 140, 0),
    "violet": (148, 0, 211),
    "rose": (255, 105, 180),
    "marron": (139, 69, 19),
    "turquoise": (64, 224, 208),
}

def closest_color_name(rgb):
    r, g, b = rgb
    best_name, best_dist = None, float("inf")
    for name, (cr, cg, cb) in BASE_COLORS.items():
        dist = (r - cr) ** 2 + (g - cg) ** 2 + (b - cb) ** 2
        if dist < best_dist:
            best_name, best_dist = name, dist
    return best_name

def extract_dominant_colors(image_path, n_colors=5):
    with Image.open(image_path) as img:
        img = img.convert("RGB")
        img = img.resize((256, int(img.height * 256 / img.width))) if img.width > 256 else img
        pixels = np.array(img).reshape(-1, 3)

    k = min(n_colors, len(pixels))
    kmeans = KMeans(n_clusters=k, n_init=5, random_state=0)
    labels = kmeans.fit_predict(pixels)
    counts = Counter(labels)
    # Trier les centres par fréquence décroissante
    ordered = sorted(counts.items(), key=lambda x: x[1], reverse=True)
    colors = [kmeans.cluster_centers_[idx].astype(int).tolist() for idx, _ in ordered]
    color_names = [closest_color_name(c) for c in colors]
    return colors, color_names

def compute_orientation(width, height, eps=0.05):
    if not width or not height:
        return "inconnu"
    ratio = width / height
    if abs(ratio - 1) <= eps:
        return "carre"
    return "paysage" if ratio > 1 else "portrait"

def size_category(width, height):
    if not width or not height:
        return "inconnu"
    max_side = max(width, height)
    if max_side < 500:
        return "vignette"
    if max_side <= 1500:
        return "moyenne"
    return "grande"

def extract_tags(meta, color_names):
    tags = set()
    if meta.get("topic"):
        tags.add(meta["topic"].lower())
    if meta.get("description"):
        words = re.findall(r"[a-zA-ZéèêàùçôîïâÊÈÉÀÙÇÔÎÏÂ']+", meta["description"].lower())
        tags.update(w for w in words if len(w) > 3)
    user = meta.get("username")
    if user:
        tags.add(f"auteur:{user}")
    tags.update(color_names)
    return list(sorted(tags))

def load_metadata():
    if not os.path.exists(METADATA_PATH):
        raise FileNotFoundError("Le fichier data/images_metadata.json est introuvable.")
    with open(METADATA_PATH, "r", encoding="utf-8") as f:
        content = f.read().strip()
        if not content:
            return []
        return json.loads(content)

def annotate_images(metadata_list, n_colors=5):
    os.makedirs("data", exist_ok=True)
    labels = {}

    for meta in metadata_list:
        filename = meta.get("file_name")
        if not filename:
            continue
        image_path = os.path.join("images", filename)
        if not os.path.exists(image_path):
            continue

        # Dimensions : priorité à ce qui est dans les métadonnées, sinon on relit l'image
        width = meta.get("dimensions", {}).get("width")
        height = meta.get("dimensions", {}).get("height")
        if not width or not height:
            try:
                with Image.open(image_path) as img:
                    width, height = img.size
            except Exception:
                width, height = None, None

        try:
            colors, color_names = extract_dominant_colors(image_path, n_colors=n_colors)
        except Exception:
            colors, color_names = [], []

        orientation = compute_orientation(width, height)
        size_cat = size_category(width, height)
        tags = extract_tags(meta, color_names)

        labels[filename] = {
            "predominant_colors": colors,
            "color_names": color_names,
            "orientation": orientation,
            "size_category": size_cat,
            "tags": tags,
        }

    with open(LABELS_PATH, "w", encoding="utf-8") as f:
        json.dump(labels, f, ensure_ascii=False, indent=2)
    return labels

# Exécution de l'étiquetage
metadata_list = load_metadata()
labels = annotate_images(metadata_list, n_colors=5)

print(f"Annotations générées pour {len(labels)} images et enregistrées dans {LABELS_PATH}")
# Aperçu rapide
for i, (fname, data) in enumerate(labels.items()):
    if i >= 3:
        break
    print(fname, "->", data)
```

```python
# Tâche 3 : Analyse des données

import random
from collections import Counter
import json
import os

import pandas as pd
from sklearn.cluster import KMeans

USERS_PATH = "data/users.json"

# 1. Simulation d'utilisateurs et de favoris

def simulate_users(labels, n_users=5, min_fav=10, max_fav=20, random_state=42):
    random.seed(random_state)
    image_ids = list(labels.keys())

    # On force n_fav à être strictement inférieur au nombre d'images
    # pour garantir qu'il reste toujours au moins une image non favorite.
    if len(image_ids) <= 1:
        raise ValueError("Il faut au moins 2 images pour simuler des favoris / non favoris.")
    max_fav = min(max_fav, len(image_ids) - 1)
    min_fav = min(min_fav, max_fav)
    topics = set()
    for lbl in labels.values():
        for tag in lbl.get("tags", []):
            if not tag.startswith("auteur:"):
                topics.add(tag)
    topics = list(topics)

    users = []
    for i in range(n_users):
        user_id = f"utilisateur_{i+1:03d}"
        n_fav = random.randint(min_fav, max_fav)
        # On force un biais de préférence en filtrant sur un topic si possible
        preferred_topic = random.choice(topics) if topics else None
        if preferred_topic:
            biased_images = [img for img, lbl in labels.items() if preferred_topic in lbl.get("tags", [])]
        else:
            biased_images = []
        remaining = [img for img in image_ids if img not in biased_images]
        favorites = []
        # 60 % des favoris dans le thème préféré, le reste aléatoire
        n_biased = int(n_fav * 0.6)
        if biased_images:
            favorites.extend(random.sample(biased_images, min(n_biased, len(biased_images))))
        remaining_candidates = [img for img in remaining if img not in favorites]
        if remaining_candidates:
            n_remaining = n_fav - len(favorites)
            favorites.extend(random.sample(remaining_candidates, min(n_remaining, len(remaining_candidates))))

        users.append({
            "user_id": user_id,
            "preferred_topic": preferred_topic,
            "favorite_images": favorites,
        })
    return users


# 2. Construction des profils utilisateurs

def build_user_profiles(users, labels):
    profiles = []
    for user in users:
        fav_labels = [labels[img] for img in user["favorite_images"] if img in labels]
        if not fav_labels:
            continue
        # Couleurs : flat list de tous les color_names
        color_names = []
        orientations = []
        sizes = []
        tags = []
        for lbl in fav_labels:
            color_names.extend(lbl.get("color_names", []))
            orientations.append(lbl.get("orientation"))
            sizes.append(lbl.get("size_category"))
            tags.extend(lbl.get("tags", []))
        def most_common(lst):
            lst = [x for x in lst if x]
            return Counter(lst).most_common(1)[0][0] if lst else None
        favorite_colors = [c for c, _ in Counter(color_names).most_common(5)]
        favorite_orientation = most_common(orientations)
        favorite_size = most_common(sizes)
        favorite_tags = [t for t, _ in Counter(tags).most_common(10)]

        profile = {
            "user_id": user["user_id"],
            "preferred_topic": user.get("preferred_topic"),
            "favorite_colors": favorite_colors,
            "favorite_orientation": favorite_orientation,
            "favorite_size": favorite_size,
            "favorite_tags": favorite_tags,
            "favorite_images": user["favorite_images"],
        }
        profiles.append(profile)
    return profiles


# 3. Analyse globale des tendances

def analyze_preferences(profiles):
    # DataFrame pour faciliter les analyses
    rows = []
    for p in profiles:
        for img in p["favorite_images"]:
            rows.append({
                "user_id": p["user_id"],
                "preferred_topic": p.get("preferred_topic"),
                "favorite_colors": p["favorite_colors"],
                "favorite_orientation": p["favorite_orientation"],
                "favorite_size": p["favorite_size"],
                "favorite_tags": p["favorite_tags"],
            })
    df = pd.DataFrame(rows)
    if df.empty:
        print("Aucune donnée utilisateur à analyser.")
        return None, None

    # Popularité globale des couleurs
    all_colors = [c for sub in df["favorite_colors"] for c in sub]
    color_counts = Counter(all_colors)
    # Popularité globale des tags
    all_tags = [t for sub in df["favorite_tags"] for t in sub]
    tag_counts = Counter(all_tags)

    print("Couleurs les plus populaires globalement :")
    for color, cnt in color_counts.most_common(10):
        print(f"  {color}: {cnt}")

    print("\nTags les plus fréquents :")
    for tag, cnt in tag_counts.most_common(10):
        print(f"  {tag}: {cnt}")

    # Clustering simple des utilisateurs sur base de leurs couleurs préférées (one-hot basique)
    unique_colors = sorted(set(all_colors))
    user_vectors = []
    user_ids = []
    for p in profiles:
        vec = [1 if c in p["favorite_colors"] else 0 for c in unique_colors]
        user_vectors.append(vec)
        user_ids.append(p["user_id"])
    if len(user_vectors) >= 2:
        k = min(3, len(user_vectors))
        km = KMeans(n_clusters=k, n_init=5, random_state=0)
        clusters = km.fit_predict(user_vectors)
        print("\nClusters d'utilisateurs (basés sur les couleurs favorites) :")
        for cid in range(k):
            members = [uid for uid, cl in zip(user_ids, clusters) if cl == cid]
            print(f"  Cluster {cid}: {members}")
    else:
        print("\nPas assez d'utilisateurs pour un clustering significatif.")

    return color_counts, tag_counts


# Chargement des labels (Tâche 2)
with open("data/images_labels.json", "r", encoding="utf-8") as f:
    labels = json.load(f)

# Exécution complète de la Tâche 3
os.makedirs("data", exist_ok=True)
users = simulate_users(labels, n_users=5, min_fav=10, max_fav=20)
profiles = build_user_profiles(users, labels)

with open(USERS_PATH, "w", encoding="utf-8") as f:
    json.dump(profiles, f, ensure_ascii=False, indent=2)

print(f"Profils utilisateurs générés pour {len(profiles)} utilisateurs et enregistrés dans {USERS_PATH}")

color_counts, tag_counts = analyze_preferences(profiles)
```

```python
# Tâche 5 : Système de recommandation (Option A - SVM)

import json
from typing import List, Tuple

import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import LinearSVC

# Chargement des données nécessaires
with open("data/images_labels.json", "r", encoding="utf-8") as f:
    images_labels = json.load(f)

with open("data/users.json", "r", encoding="utf-8") as f:
    user_profiles = json.load(f)


def _build_image_corpus() -> Tuple[List[str], List[str]]:
    """Construit un 'document' texte par image à partir de ses caractéristiques.

    On concatène : tags, noms de couleurs, orientation et taille.
    Cela permet d'utiliser un CountVectorizer très simplement.
    """
    image_ids = []
    corpus = []
    for image_id, meta in images_labels.items():
        tags = meta.get("tags", [])
        colors = meta.get("color_names", [])
        orientation = [meta.get("orientation")] if meta.get("orientation") else []
        size_cat = [meta.get("size_category")] if meta.get("size_category") else []

        tokens = list(tags) + list(colors) + orientation + size_cat
        doc = " ".join(str(t) for t in tokens if t)
        image_ids.append(image_id)
        corpus.append(doc)
    return image_ids, corpus


def _train_user_svm(user_id: str) -> Tuple[LinearSVC, np.ndarray, List[str], CountVectorizer]:
    """Entraîne un classificateur SVM Favori / Non favori pour un utilisateur.

    - X : caractéristiques des images (sac de mots sur tags/couleurs/orientation/taille)
    - y : 1 si l'image est favorite pour l'utilisateur, 0 sinon
    """
    # Construire matrice de caractéristiques pour toutes les images
    image_ids, corpus = _build_image_corpus()
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(corpus)

    # Récupérer les favoris de l'utilisateur
    profile = next(p for p in user_profiles if p["user_id"] == user_id)
    fav_set = set(profile.get("favorite_images", []))

    # Créer les étiquettes (1 = favori, 0 = non favori)
    y = np.array([1 if img in fav_set else 0 for img in image_ids])

    # Il faut au moins une image positive ET une négative
    n_pos = int(y.sum())
    n_neg = int(len(y) - n_pos)
    if n_pos == 0 or n_neg == 0:
        raise ValueError("L'utilisateur doit avoir au moins une image favorite et une non favorite pour entraîner le modèle.")

    # Entraînement SVM linéaire binaire
    clf = LinearSVC()
    clf.fit(X, y)

    # On renvoie aussi X et image_ids pour réutiliser les scores
    return clf, X, image_ids, vectorizer

def explain_image(X, img_index, weights, feature_names, top_k=3):
    """Retourne les top features (tokens) qui poussent l'image vers 'favori'."""
    x_row = X[img_index]
    present_idx = x_row.nonzero()[1]
    if len(present_idx) == 0:
        return []

    present_weights = weights[present_idx]

    # On ne garde que les poids positifs (ce qui augmente la probabilité 'favori')
    pos_mask = present_weights > 0
    present_idx = present_idx[pos_mask]
    present_weights = present_weights[pos_mask]
    if len(present_idx) == 0:
        return []

    sorted_idx = np.argsort(-present_weights)
    top_idx = sorted_idx[:top_k]

    top_features = feature_names[present_idx[top_idx]]
    top_weights = present_weights[top_idx]

    return [(str(f), float(w)) for f, w in zip(top_features, top_weights)]


def recommend_images(user_id: str, n_recommendations: int = 5) -> List[Tuple[str, str]]:
    """Recommande des images à un utilisateur via un SVM basé contenu.

    Le SVM est entraîné sur : Favori(1) / Non favori(0).
    Les 'raisons' retournées sont les tokens (tags/couleurs/orientation/taille)
    avec les plus gros poids positifs présents dans l'image.
    """
    try:
        clf, X, image_ids, vectorizer = _train_user_svm(user_id)
    except ValueError as e:
        print(f"Impossible d'entraîner le modèle pour {user_id} : {e}")
        return []

    profile = next(p for p in user_profiles if p["user_id"] == user_id)
    fav_set = set(profile.get("favorite_images", []))

    scores = clf.decision_function(X)
    ranked = sorted(zip(image_ids, scores), key=lambda t: t[1], reverse=True)

    id_to_idx = {img_id: i for i, img_id in enumerate(image_ids)}
    feature_names = np.array(vectorizer.get_feature_names_out())
    weights = clf.coef_[0]

    recommendations: List[Tuple[str, str]] = []

    for img_id, score in ranked:
        if img_id in fav_set:
            continue

        img_index = id_to_idx[img_id]
        reasons = explain_image(X, img_index, weights, feature_names, top_k=3)

        if reasons:
            reason_text = ", ".join([f"{tok} ({w:.2f})" for tok, w in reasons])
        else:
            reason_text = "car elle partage des caractéristiques avec vos favoris"

        recommendations.append((img_id, reason_text))
        if len(recommendations) >= n_recommendations:
            break

    return recommendations

for i in range(len(user_profiles)):
    user = user_profiles[i]["user_id"]
    reco = recommend_images(user, n_recommendations=5)
    print(f"\nRecommandations pour {user} :")
    for img, why in reco:
        print(f"- {img} -> {why}")
```
